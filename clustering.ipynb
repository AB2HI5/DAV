```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=pd.read_csv("/content/Mall_Customers-1.csv")
df.head()
```

```python
from sklearn.cluster import KMeans
x=df[['Age','Spending Score (1-100)']].copy()
```

```python
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
  kmeans.fit(x)
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('Slecting the Number of Clusters using the Elbow Method')
plt.xlabel('Cluster')
plt.ylabel('wcss')
plt.show()
```

```python
for k in range(1,11):
  plt.scatter(x['Age'],x['Spending Score (1-100)'])
  plt.title("cluster identified by k-means clustering")
plt.ylabel("spending score (1-100)")
plt.xlabel("Age")
plt.show()
```
