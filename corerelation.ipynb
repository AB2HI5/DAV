```python
from scipy import stats
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,81,88,111,86,103,87,94,78,77,85,86]
```

```python
slope, intercept, r, p, std_err = stats.linregress(x,y)
```

```python
def myfun(x):
  return slope*+intercept
```

```python
mymodel=list(map(myfunc,x))
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,mymodel)
```
